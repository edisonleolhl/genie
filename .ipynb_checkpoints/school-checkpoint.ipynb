```python
import pandas as pd
import jieba
from snownlp import SnowNLP
```

```python
df = pd.read_csv("whisper.csv", encoding="utf-8")
df['content'] = df['content'].astype(str)
```

```python
df.shape
```

```python
df.tail()
```

```python
def my_sentiments(text):
    return SnowNLP(text).sentiments
```

```python
df["sentiments"] = df.content.apply(my_sentiments)
```

```python
df['date'] = pd.to_datetime(df['publishTime']) #将数据类型转换为日期类型
df["2017-01"].head()
```

```python
print(df["2016-11"]["sentiments"].mean())
print(df["2016-11"]["sentiments"].std()) # 样本标准差（无偏）
```

```python
df["dayofweek"] = df['date'].apply(lambda x:x.weekday())
df.set_index("dayofweek")
```

```python
dayofweeks = list(range(7))
sentiments_dayofweek_trend = []
for dayofweek in dayofweeks:    
    sentiments_dayofweek_trend.append(df[df['dayofweek'] == dayofweek].sentiments.mean())
sentiments_dayofweek_trend
```

```python
months = ["2016-11", "2016-12", 
          "2017-01", "2017-02", "2017-03", "2017-04", "2017-05", "2017-06",
          "2017-07", "2017-08", "2017-09", "2017-10", "2017-11",]
sentiments_mean_trend = []
sentiments_std_trend = []
whisper_monthly_number = []
for month in months:
    sentiments_mean_trend.append(df[month]["sentiments"].mean())
    sentiments_std_trend.append(df[month]["sentiments"].std())
    whisper_monthly_number.append(df[month]["content"].count())
print(sentiments_mean_trend)
print(sentiments_std_trend)
print(whisper_monthly_number)
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
```

```python
months
```

```python
xs = [datetime.strptime(d, '%Y-%m').date() for d in months]
xs
```

```python
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.rcParams.update({'font.size': 15})

# Plot mean
plt.figure(figsize=(22,10))
plt.ylim((0.7, 0.9))
plt.yticks([0.7, 0.9],['$BAD\ (mean=0.7)$', '$GOOD\ (mean=0.9)$'])
plt.plot(xs, sentiments_mean_trend, linewidth=1.5)
plt.xlabel(u"月份 (2016年11月 -- 2017年11月)", fontproperties='SimHei')
plt.ylabel(u"情绪指数平均值", fontproperties='SimHei')
plt.title(u"悄悄话情绪指数（每月平均）的时序变化", fontproperties='SimHei')
plt.savefig('D:\genie\whisper_time_varient_mean.png', dpi=100, pad_inches=10)
plt.show()

# Plot std
plt.figure(figsize=(22,10))
plt.ylim((0.3, 0.5))
plt.yticks([0.3, 0.5],['$STABLE\ (std=0.3)$', '$UNSTABLE\ (std=0.5)$'])
plt.plot(xs, sentiments_std_trend, linewidth=1.5)
plt.xlabel(u"月份 (2016年11月 -- 2017年11月)", fontproperties='SimHei')
plt.ylabel(u"情绪指数标准差", fontproperties='SimHei')
plt.title(u"悄悄话情绪指数（每月标准差）的时序变化", fontproperties='SimHei')
plt.savefig('D:\genie\whisper_time_varient_std.png', dpi=100, pad_inches=10)
plt.show()
```
